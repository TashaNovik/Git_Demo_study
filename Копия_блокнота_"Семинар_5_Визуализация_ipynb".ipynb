```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
import numpy as np
```

```python
# Загрузим датасет iris из seaborn
iris = sns.load_dataset("iris")
```

**Датасет iris** — это классический набор данных, который используется для задач классификации и визуализации. Он содержит информацию о трех различных видах цветков ириса (Setosa, Versicolor и Virginica), а также их физических характеристиках. Датасет включает 150 наблюдений и 5 переменных.

Описание переменных:
- sepal_length: Длина чашелистика (в сантиметрах).
- sepal_width: Ширина чашелистика (в сантиметрах).
- petal_length: Длина лепестка (в сантиметрах).
- petal_width: Ширина лепестка (в сантиметрах).
- species: Вид ириса, представленный тремя категориями — setosa, versicolor и virginica.

```python
iris.head()
```

### Line Plot: Линейный график длины и ширины чашелистика по индексам.

```python
# 1. Линейный график (Line Plot)
plt.figure(figsize=(8, 5))
plt.plot(iris.index, iris["sepal_length"], color="b", label="Sepal Length")
plt.plot(iris.index, iris["sepal_width"], color="g", label="Sepal Width")
plt.title("Line Plot")
plt.xlabel("Index")
plt.ylabel("Length (cm)")
plt.legend()
plt.show()
```

### Bar Plot: Столбчатая диаграмма с количеством образцов каждого вида ириса.

```python
# 2. Столбчатая диаграмма (Bar Plot)
species_counts = iris['species'].value_counts()
plt.figure(figsize=(8, 5))
plt.bar(species_counts.index, species_counts.values, color=["purple", "orange", "cyan"])
plt.title("Bar Plot")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```

### Histogram: Гистограмма распределения длины лепестков

```python
# 3. Гистограмма (Histogram)
plt.figure(figsize=(8, 5))
plt.hist(iris["petal_length"], bins=20, color="salmon", edgecolor="black")
plt.title("Histogram of Petal Length")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")
plt.show()
```

### Scatter Plot: Точечная диаграмма связи между длиной чашелистика и длиной лепестка, окраска по ширине лепестка.

```python
# 4. Scatter Plot (Точечная диаграмма)
plt.figure(figsize=(8, 5))
plt.scatter(iris["sepal_length"], iris["petal_length"], c=iris["petal_width"], cmap="viridis", edgecolor="k", s=50)
plt.title("Scatter Plot")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.colorbar(label="Petal Width (cm)")
plt.show()
```

### Box Plot: Коробчатая диаграмма для распределения длины чашелистика по видам.

```python
# 5. Boxplot
plt.figure(figsize=(8, 5))
sns.boxplot(x="species", y="sepal_length", data=iris, palette="Set3")
plt.title("Box Plot of Sepal Length by Species")
plt.xlabel("Species")
plt.ylabel("Sepal Length (cm)")
plt.show()
```

### Heatmap: Тепловая карта корреляций между числовыми переменными.

```python
# 6. Heatmap (Тепловая карта)
plt.figure(figsize=(8, 5))
corr = iris.drop(columns="species").corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Heatmap of Correlations")
plt.show()
```

### Многофигурные графики

```python
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

# 1. Линейный график (Line Plot)
axs[0, 0].plot(iris.index, iris["sepal_length"], color="b", label="Sepal Length")
axs[0, 0].plot(iris.index, iris["sepal_width"], color="g", label="Sepal Width")
axs[0, 0].set_title("Line Plot")
axs[0, 0].set_xlabel("Index")
axs[0, 0].set_ylabel("Length (cm)")
axs[0, 0].legend()

# 2. Столбчатая диаграмма (Bar Plot)
species_counts = iris['species'].value_counts()
axs[0, 1].bar(species_counts.index, species_counts.values, color=["purple", "orange", "cyan"])
axs[0, 1].set_title("Bar Plot")
axs[0, 1].set_xlabel("Species")
axs[0, 1].set_ylabel("Count")

# 3. Гистограмма (Histogram)
axs[0, 2].hist(iris["petal_length"], bins=20, color="salmon", edgecolor="black")
axs[0, 2].set_title("Histogram of Petal Length")
axs[0, 2].set_xlabel("Petal Length (cm)")
axs[0, 2].set_ylabel("Frequency")

# 4. Scatter Plot (Точечная диаграмма)
axs[1, 0].scatter(iris["sepal_length"], iris["petal_length"], c=iris["petal_width"], cmap="viridis", edgecolor="k", s=50)
axs[1, 0].set_title("Scatter Plot")
axs[1, 0].set_xlabel("Sepal Length (cm)")
axs[1, 0].set_ylabel("Petal Length (cm)")

# 5. Boxplot (Коробчатая диаграмма)
sns.boxplot(x="species", y="sepal_length", data=iris, ax=axs[1, 1], palette="Set3")
axs[1, 1].set_title("Box Plot of Sepal Length by Species")
axs[1, 1].set_xlabel("Species")
axs[1, 1].set_ylabel("Sepal Length (cm)")

# 6. Heatmap (Тепловая карта)
corr = iris.drop(columns="species").corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", ax=axs[1, 2])
axs[1, 2].set_title("Heatmap of Correlations")

# Отображаем графики
plt.tight_layout()
#plt.savefig("iris_dashboard.png", dpi=300)
plt.show()
```
